```r
# read the file
df <- read.csv("amazon-dataset.csv")

# remove the last column
df <- df[,-c(23)]

# personalize data
df$Ordering.Customer.Email = 'scottdownie0506@gwu.edu'
df$Shipping.Address.Name = 'Scott'
df$Buyer.Name = 'Scott'


# change prices to numerical values for summary 
df$Subtotal <- sub("\\$", '', df$Subtotal)
df$Subtotal <- as.numeric(df$Subtotal)
df$Shipping.Charge <- sub("\\$", '', df$Shipping.Charge)
df$Shipping.Charge <- as.numeric(df$Shipping.Charge)
df$Tax.Before.Promotions <- sub("\\$", '', df$Tax.Before.Promotions)
df$Tax.Before.Promotions <- as.numeric(df$Tax.Before.Promotions)
df$Total.Promotions <- sub("\\$", '', df$Total.Promotions)
df$Total.Promotions <- as.numeric(df$Total.Promotions)
df$Tax.Charged <- sub("\\$", '', df$Tax.Charged)
df$Tax.Charged <- as.numeric(df$Tax.Charged)
df$Total.Charged <- sub("\\$", '', df$Total.Charged)
df$Total.Charged <- as.numeric(df$Total.Charged)

# call data frame
df
```

```r
# different statistics 

MeanTotal <- mean(df$Total.Charged)
MaxTotal <- max(df$Total.Charged)
MinTotal <- min(df$Total.Charged)
MedianTotal <- median(df$Total.Charged)

MaxPromotions <- max(df$Total.Promotions)

MaxShipping <- max(df$Shipping.Charge)
MeanShipping <- mean(df$Shipping.Charge)

print(paste("The average total for each order was $", MeanTotal, sep=''))
print(paste("The maximum total for an order was $", MaxTotal, sep=''))
print(paste("The lowest amount I spent on an order was $", MinTotal, sep=''))
print(paste("The median order price was $", MedianTotal, sep=''))
print('')
print(paste("$", MaxPromotions, " was the most promotions I received", sep=''))
print('')
print(paste("The average shipping price for each order was $", MeanShipping, sep=''))
print(paste("$", MaxShipping, " was the most I paid for shipping", sep=''))
```

```r
print('This is a graph of the amount I paid for each order over time.')

plot(df$Total.Charged, ylab = 'Amount Paid', xlab = 'Order Number', main = 'Amount Paid per Order', pch=15, col = 'red')
```

```r
x1 <- 0
x2 <- 0
x3 <- 0

for (x in df$Shipping.Charge) {
    if (x == 0) {
        x1 <- x1 + 1
    } else if (x > 5) {
        x3 <- x3 + 1
    } else {
        x2 <- x2 + 1
    }
}

colors <- c('red', 'blue', 'green')
pies <- c(x1, x2, x3)
labels <- c('Free Shipping', 'Less than $5', 'More than $10')
pie(pies, label = labels, main = 'Shipping Prices', col = colors)
```
